```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier # Import Decision Tree Classifier
from sklearn.model_selection import train_test_split # Import train_test_split function
from sklearn.metrics import accuracy_score
from sklearn.externals import joblib
from sklearn.preprocessing import QuantileTransformer, LabelEncoder
```

```python
dataset = pd.read_csv("irrigation_data.csv")
```

```python
categorical_features_names = ['temp', 'humidity', 'weather', 'wind', 'label']
one_hot = pd.get_dummies(dataset['weather'])
dataset = dataset.join(one_hot)
dataset = dataset.drop('weather', axis = 1)
```

```python
dataset
```

```python
X = dataset.drop('label', axis=1)
```

```python
X
```

```python
y = dataset['label']
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20)
```

```python
classifier = DecisionTreeClassifier()
classifier.fit(X_train, y_train)
```

```python
y_pred = classifier.predict(X_test)
acc = accuracy_score(y_test, y_pred)
print("Accuracy of classifier: " + str(acc))
```
